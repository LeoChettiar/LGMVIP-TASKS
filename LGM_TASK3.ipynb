**GRAPHICAL** **REPRESENTATION**

MONTH WISE ATTACKS

```python
attacks=data["Month"].value_counts(dropna=False).sort_index().to_frame().reset_index().rename(columns={"index":"Month","Month":"Attacks"}).set_index("Month")
attacks.head(13)
```

CONCLUSION-IT SHOWS THE NUMBER OF ATTACKS IN EACH MONTH

```python
attacks.plot(kind="bar",color="blue",figsize=(15,6),fontsize=13)
plt.title("Timeline of Attacks",fontsize=15)
plt.xlabel("Month",fontsize=15)
plt.ylabel("Number of Attacks",fontsize=15)
plt.show()
```

**Conclusion**-It shows thw number of attacks in each month ,as shown in fig we can conclude that in 5th month there are highest number of attacks

**YEAR WISE ATTACK**

```python
attacks=data["Year"].value_counts(dropna=False).sort_index().to_frame().reset_index().rename(columns={"index":"Year","Year":"Attacks"}).set_index("Year")
attacks.head(10)
```

```python
attacks.plot(kind="bar",color="green",figsize=(15,6),fontsize=13)
plt.title("Timeline of Attacks",fontsize=15)
plt.xlabel("Year",fontsize=15)
plt.ylabel("Number of Attacks",fontsize=15)
plt.show()
```

**Conclusion**-from the above fig we can colclude that highest number of attack was happened in the year 1979,1980,1981

```python
sns.countplot(x='Region', data=data)
plt.xticks(rotation=90)
plt.xlabel('Region of Attacked')
plt.ylabel('No. of Attack')
plt.title('Most Attacked region')
plt.show()
```

**Conclusion**-from the above fig we can conclude that in western europe and south america are the most attacked region

```python
sns.countplot(x='Target Type',data=data,palette='Spectral')
plt.xticks(rotation=90)
plt.xlabel('target type')
plt.ylabel('No. of Attack')
plt.title('Target of Terrorist')
plt.show()
```

**Conclusion**-The above fig shows the target of terrorist.

```python
grp=data["Group Name"].value_counts()[1:10]
grp
```

**GROUP** **WISE** **ATTACK**

```python
grp.plot(kind="bar",color="blue",figsize=(15,6))
plt.title("Group wise Attacks",fontsize=13)
plt.xlabel("Terrorist Groups",fontsize=13)
plt.xticks(fontsize=12)
plt.ylabel("Number of Attacks",fontsize=13)
plt.show()
```

**Conclusion**:-from above graph we can conclude that how many number of attacks was performed by each terrorist group

**REGION** **WISE** **ATTACK**

```python
plt.figure(figsize=(10,8))
plt.title('Region wise Attack')
sns.countplot(x=data['Region'],palette='PuRd_r',order=data['Region'].value_counts()[0:7].index)
plt.xticks(rotation=90)
```
