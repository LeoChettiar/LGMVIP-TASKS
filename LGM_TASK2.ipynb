```python
import cv2
from google.colab.patches import cv2_imshow
import matplotlib.pyplot as plt
from google.colab.patches import *
```

**Loading The Original Image**

```python
img = cv2.imread("/content/the goats.jpg")
cv2_imshow(img)
```

**Converting The Image Into Greyscale**

```python
grey_image=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
cv2_imshow(grey_image)
```

**Converting The Image Into Inverted Image**

```python
inverted_img=255-grey_image
cv2_imshow(inverted_img)
```

**Converting The Image Into Gaussian Blur**

```python
blurred=cv2.GaussianBlur(inverted_img,(21,21),0)
cv2_imshow(blurred)
```

**Converting The Image Into Inverted Blurred**

```python
inverted_blurred=255-blurred
cv2_imshow(inverted_blurred)
```

**Converting The Image Into Sketch**

```python
pencil_sketch=cv2.divide(grey_image,inverted_blurred,scale=256.0)
cv2_imshow(pencil_sketch)
```

**Final Output**

```python
plt.figure(figsize=(20,20))
plt.subplot(1,5,1)
plt.imshow(img)
plt.axis("off")
plt.title("Original Image")
plt.subplot(1,5,2)
plt.imshow(grey_image,cmap="gray")
plt.axis("off")
plt.title("GrayScale Image")
plt.subplot(1,5,3)
plt.imshow(inverted_img,cmap="gray")
plt.axis("off")
plt.title("Inverted Image")
plt.subplot(1,5,4)
plt.imshow(blurred,cmap="gray")
plt.axis("off")
plt.title("Blurred Image")
plt.subplot(1,5,5)
plt.imshow(inverted_blurred,cmap="gray")
plt.axis("off")
plt.title("Inverted Blurred Image")
plt.imshow(pencil_sketch,cmap="gray")
plt.axis("off")
plt.title("Final Pencil Sketch Image")
plt.show()
```
